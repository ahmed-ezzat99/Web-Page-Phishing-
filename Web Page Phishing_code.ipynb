# Model Building Stage

```python
X = data.iloc[:,:-1]   #features
y = data.iloc[:, -1]   #label 
```

```python
## Importing Libraries for Machine Learning model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score

```

```python
# Splitting the Dataset into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
#Training Data Information

print("Trainig Data..")
print("The shape of training features: ", X_train.shape)
print("The shape of training labels: ", y_train.shape)
```

```python
#Testing Data Information

print("Testing Data..")
print("The shape of testing features: ", X_test.shape)
print("The shape of testing labels: ", y_test.shape)
```

```python
#feature scaling
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_S = scaler.transform(X_train)
X_test_S = scaler.transform(X_test)
```

```python
#Support Vector Machine (SVM) Classifier Initialization

from sklearn.svm import SVC
model = SVC()
```

```python
#Fitting the SVM Classifier to the scaled Training Data

model.fit(X_train_S,y_train)
```

```python
#Predicting Labels for Testing Data using SVM Classifier
y_pred = model.predict(X_test_S)
```

# Model Evaluation

```python
#Calculating Accuracy 
accuracy = accuracy_score(y_test, y_pred)* 100
```

```python
print("The Accuracy = ", accuracy, "%")
```

```python
#Calculating Error
error_rate = (1 - accuracy_score(y_test, y_pred))*100
print("Error Rate:", error_rate , "%")
```
